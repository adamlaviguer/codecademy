## Project Data

```python
import csv

def read_file(file):
    with open(file) as data_file:
        data_dict = csv.DictReader(data_file)
        data_list = list(data_dict)
        return data_list

data_list = read_file('insurance.csv')

num_records = len(data_list)
print("Total Records: {}".format(num_records))

min_age, max_age, min_children, max_children = 0, 0, 0, 0
for record in data_list:
    current_age = int(record["age"])
    if min_age == 0:
        min_age = current_age
    elif current_age < min_age:
        min_age = current_age
    if max_age == 0:
        max_age = current_age
    elif current_age > max_age:
        max_age = current_age

    current_children = int(record["children"])
    if min_children == 0:
        min_children = current_children
    elif current_children < min_children:
        min_children = current_children
    if max_children == 0:
        max_children = current_children
    elif current_children > max_children:
        max_children = current_children
print("Max Age: {}".format(max_age))
print("Min Age: {}".format(min_age))
print("Max Children: {}".format(max_children))
print("Min Children: {}".format(min_children))
```

#### 1. Define demographics

```python
from itertools import product

# Define the lists
ages = [(18, 21), (22, 35), (36,50), (51, 64)]
sex = ['male', 'female']
children = ['yes', 'no']
smoker = ['yes', 'no']
region = ['southeast', 'southwest', 'northeast', 'northwest']

# Create all possible combinations
combinations = list(product(ages, sex, children, smoker, region))

# Define the column names
columns = ['Age', 'Sex', 'Children', 'Smoker', 'Region']

# Create the dictionary with unique integer IDs
demo_dict = {i: dict(zip(columns, combo)) for i, combo in enumerate(combinations, start=1)}

# Optional: print a sample
print(f"Total combinations: {len(demo_dict)}")
print("Sample entry:", demo_dict[1])
```

#### 2. A. What is the average medical charge overall?

```python
total_charges = 0
for data in data_list:
    total_charges += float(data["charges"])
print("The dataset contains a total charge amount of {} dollars. Across the {} records, the average charges are {} dollars.".format(round(total_charges, 2), num_records, round(total_charges/num_records, 2)))
```

#### 2. B. What is the distribution of BMI across different age groups?

```python
age_group_data = {ages[0]: [], ages[1]: [], ages[2]: [], ages[3]: []}
ages_0 = [i for i in range(ages[0][0], ages[0][1]+1)]
ages_1 = [i for i in range(ages[1][0], ages[1][1]+1)]
ages_2 = [i for i in range(ages[2][0], ages[2][1]+1)]
ages_3 = [i for i in range(ages[3][0], ages[3][1]+1)]

for data in data_list:
    current_age = int(data["age"])
    if current_age in ages_0:
        age_group_data[ages[0]].append(data)
    elif current_age in ages_1:
        age_group_data[ages[1]].append(data)
    elif current_age in ages_2:
        age_group_data[ages[2]].append(data)
    elif current_age in ages_3:
        age_group_data[ages[3]].append(data)

age_group_data_flat = []
for data in data_list:
    current_age = int(data["age"])
    if current_age in ages_0:
        age_group_data_flat.append({"Age Group": str(ages[0]), "BMI": float(data["bmi"])})
    elif current_age in ages_1:
        age_group_data_flat.append({"Age Group": str(ages[1]), "BMI": float(data["bmi"])})
    elif current_age in ages_2:
        age_group_data_flat.append({"Age Group": str(ages[2]), "BMI": float(data["bmi"])})
    elif current_age in ages_3:
        age_group_data_flat.append({"Age Group": str(ages[3]), "BMI": float(data["bmi"])})
# print(age_group_data_flat)

import pandas as pd
import plotly.express as px

# Create a DataFrame
df = pd.DataFrame(age_group_data_flat)

# Create a box plot to visualize BMI distribution across age groups
fig = px.box(df, x='Age Group', y='BMI', points='all',
             labels={'Age Group': 'Age Group', 'BMI': 'BMI'},
             title='Distribution of BMI Across Age Groups')

# Show the plot
fig.show()
```

The distributions shown above do not indicate that BMI is correlated to the age group.

#### 2. C. What percentage of individuals are smokers vs. non-smokers?

```python
num_smokers = 0
for data in data_list:
    if data["smoker"] == "yes":
        num_smokers += 1
num_non_smokers = num_records - num_smokers
smoker_percent = (num_smokers / num_records) * 100
print("There are {} smokers in the dataset which makes up {}% of the records.".format(num_smokers, round(smoker_percent, 1)))
print("The remaining {} non-smokers make up {}% of the records.".format(num_non_smokers, round(100-smoker_percent, 1)))
```
